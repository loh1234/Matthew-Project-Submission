# Intermediate Level

```python
# Count how many movies have each of the content ratings.
movies.content_rating.value_counts()
```

```python
# Use a visualization to display that same data, including a title and x and y labels.
movies.content_rating.value_counts().plot(kind='bar')
plt.xlabel('Content Rating');
plt.ylabel('No. of Movies');
```

```python
# Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.
movies.content_rating.replace('NOT RATED', 'UNRATED', inplace=True)
movies.content_rating.replace('APPROVED', 'UNRATED', inplace=True)
movies.content_rating.replace('PASSED', 'UNRATED', inplace=True)
movies.content_rating.replace('GP', 'UNRATED', inplace=True)
```

```python
# Convert the following content ratings to "NC-17": X, TV-MA
movies.content_rating.replace('X', 'NC-17', inplace=True)
movies.content_rating.replace('TV-MA', 'NC-17', inplace=True)
movies.content_rating.value_counts()
```

```python
# Count the number of missing values in each column
print((movies.isnull().sum()))
```

```python
# If there are missing values: examine them, then fill them in with "reasonable" values.
movies[movies.content_rating.isnull()] # Displays rows with null values.
movies.content_rating.fillna(value='UNRATED', inplace=True) # replace all values with UNRATED. How to customise replacement of indivudal NaN value?
print((movies.isnull().sum()))
```

```python
# Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours
# Answer : More than 2hr = 7.95 star rating, Less than 2hr = 7.84 star rating
star_more2hr = movies.duration > 120
star_less2hr = movies.duration < 120
print(movies[star_more2hr].star_rating.mean())
print(movies[star_less2hr].star_rating.mean())
```

```python
# Use a visualization to detect whether there is a relationship between duration and star rating
movies.boxplot(column='duration', by='star_rating')
# Answer : No conclusion.
```

```python
# Calculate the average duration for each genre.
movies.genre.value_counts() # List out all the subcategories for this feature.
movies.groupby('genre')['duration'].mean() # Group the subcategories and apply function on another feature.
```

# Advanced Level

```python
# Visualize the relationship between content rating and duration.
movies.boxplot(column='duration', by='content_rating')
# Answer : PG 13 has longer movies.
```

```python
# Determine the top rated movie (by star rating) for each genre.
movies.groupby('genre')['star_rating'].max()
```

```python
# Check if there are multiple movies with the same title, and if so, determine if they are actually duplicates.
movies['is_duplicated'] = movies.duplicated(['title','duration','genre']) # Create a new column to add all entries with same title, duration and genre into it.
movies['is_duplicated'].sum() 
# Answer : Although movies have the same title, they have different duration and genres.
```

```python
# Calculate the average star rating for each genre, but only include genres with at least 10 movies

new_genre = movies.genre.value_counts() > 10 # Create new column of genres with >10 counts

movies.groupby('new_genre').['star_rating'].mean() # Find the average of the star rating for the shortlisted genres
```
