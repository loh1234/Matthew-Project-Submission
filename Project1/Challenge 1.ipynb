```python
#Step 1: Define the range of numbers to obtain the palindrome.

x = 999*999
print(x)

y = 100*100
print(y)

test_list = list(range(10000,998002))
```

```python
#Step 2: Convert the list of integers to strings.
test_list_str = [str(item) for item in test_list]
```

```python
#Step 3: Create a reverse function to reverse a string.
def reverse(strings):
    return strings[::-1]
```

```python
#Step 4: Check and print all palindromes within the list.
for strings in test_list_str:
    if (strings == reverse(strings)):
        print(strings)
```
